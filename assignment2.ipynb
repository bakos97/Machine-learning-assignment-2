```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.linalg import svd

# loads the dataset 
filename = 'heart_failure_clinical_records_dataset.csv'
df = pd.read_csv(filename)
```

```python
# checking if the features are normal distibuted
df.skew()
```

From the results above we can determine if the features follows a gaussian distribution. 
The variables with skewness > 1 are highly positively skewed.
The variables with skewness < -1 are highly negatively skewed.
The variables with 0.5 < skewness < 1 such as are moderately positively skewed.
The variables with -0.5 < skewness < -1  moderately negatively skewed.
And, the variables with -0.5 < skewness < 0.5 are symmetric i.e normally distributed 

```python
# check if its normal distibutin with histogram
df.creatinine_phosphokinase.hist()
plt.title('creatinine_phosphokinase');
```

```python
#function to return plots for the feature
#importing necessary libraries
import scipy.stats as stats
import pylab

def normality(data,feature):
    plt.figure(figsize=(10,5))
    # plt.subplot(1,2,1)
    # data.feature.hist()
    # plt.subplot(1,2,2)
    stats.probplot(data[feature],plot=pylab)
    plt.show()
```

```python
normality(df,'creatinine_phosphokinase')
```
